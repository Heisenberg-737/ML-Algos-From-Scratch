# **Exp 10: Create a classifier using multilayer perceptron on any data of your choice. Analyze the performance.**

## **Importing Libraries**

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.datasets import load_iris
from sklearn.metrics import classification_report, confusion_matrix
```

## **Training and fitting the MLP model**

```python
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(3, 3), random_state=1)
model.fit(X_train, y_train)
```

## **Predictions from trained model**

```python
pred = model.predict(X_test)
print(pred)
```

## **Model Evaluation**

```python
print('The accuracy of the Multi-layer Perceptron is:', metrics.accuracy_score(pred, y_test))
print("Confusion Matrix:")
print(confusion_matrix(y_test,pred))
print("Classification Report:")
print(classification_report(y_test, pred))
```
