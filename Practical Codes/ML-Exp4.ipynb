### **Defining Matrices**

```python
import numpy as np

A = np.random.normal(200, 15, 500) #(Concentrate around, standard deviation, no of values)
print(A)
```

## **Creating a simple histogram**

```python
import matplotlib.pyplot as plt

plt.hist(A)
plt.show()
```

## **Creating a simple pie chart**

```python
B = np.array([30, 20, 25, 15, 5, 3, 2])
mylabels = ["India", "New Zealand", "Pakistan", "Australia", "England", "Sri Lanka", "USA"]

plt.pie(B, labels=mylabels)
plt.show() 
```

## **Sine function plot**

```python
X = np.arange(0,4*np.pi,0.2)   # start,stop,step
Y = np.sin(X)

plt.plot(X,Y)
plt.show()
```

## **Cosine function plot**

```python
X = np.arange(0,4*np.pi,0.2)   # start,stop,step
Y = np.cos(X)

plt.plot(X,Y)
plt.show()
```
