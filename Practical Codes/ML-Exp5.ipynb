## **Loading Boston Housing Dataset**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston

boston_dataset = load_boston()

A = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
B = pd.Series(boston_dataset.target)
A.head()
```

## **Standardizing the data**

```python
A = (A-A.mean())/A.std()
B = (B-B.mean())/B.std()
```

## **Linear Regression using Gradient Descent**

```python
X = A['RM'] #Most close to linear relationship with B
Y = B

plt.scatter(X,Y)
plt.show()
```

## **Gradient Descent function**

```python
def grad_descent(X,Y):
    m = 0 #slope
    c = 0 #intercept

    L = 0.0001  #Learning Rate
    epochs = 1000  # The number of iterations to perform gradient descent

    n = float(len(A)) # Number of elements in X

    # Performing Gradient Descent 
    for i in range(epochs): 
        Y_pred = m*X + c  # The current predicted value of Y
        D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
        D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
        m = m - L * D_m  # Update m
        c = c - L * D_c  # Update c
    
    return (m, c)
```

```python
slope = grad_descent(X,Y)[0]
intercept = grad_descent(X,Y)[1]

print("The slope of the line of regression:", slope)
print("The intercept of the line of regression:", intercept)
```

## **Making predictions**

```python
Y_pred = slope*X + intercept

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```
