## **Defining Matrices**

```python
import numpy as np

A = np.array([[1,2,-7],[-3,4,-2],[-6,-3,9]])
B = np.array([[1,-5,3],[-2,-4,-6],[8,5,-4]])
```

## **Transpose of matrix**

```python
print("Transpose of Matrix A:")
print(A.transpose())

print("Transpose of Matrix B:")
print(B.transpose())
```

## **Adding of matrices**

```python
print("Addtion of Matrix A and B:")
print(A+B)
```

## **Subtraction of matrices**

```python
print("Subtraction of Matrix A and B:")
print(A-B)
```

## **Multiplication of matrices**

```python
print("Multiplication of Matrix A and B:")
print(A*B)
```
